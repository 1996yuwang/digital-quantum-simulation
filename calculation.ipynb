```python
from encoding import *
import numpy as np 
import scipy.sparse.linalg
from fermi_sim import *
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt
from Hamiltonian_topo import*
from scipy import sparse
```

```python
#the calculation in this notebook was done WITHOUT the encoding, 
#just to check whether we can observe the topological chracteristics

def imaginary_time_evolution(ψ, H):
    dτ = 0.1
    nsteps = 1001
    ϵmin ,vmin = sparse.linalg.eigsh(H, which = 'SA', k =1)
    # energy expectation values
    ϵ_traj = np.zeros(nsteps)
    ϵ_traj[0] = np.vdot(ψ, H.dot(ψ)).real
    for n in range(1, nsteps):
        ψ = spla.expm_multiply(-dτ*H, ψ)
        ψ /= np.linalg.norm(ψ)
        ϵ_traj[n] = np.vdot(ψ, H.dot(ψ)).real
        
    print(ϵ_traj[nsteps-1])

    plt.plot(dτ*np.arange(nsteps), ϵ_traj)
    plt.plot([0, nsteps*dτ], [ϵmin, ϵmin], "--")
    plt.xlabel("τ")
    plt.ylabel(r"$\langle \psi(\tau) | H | \psi(\tau) \rangle$")
    plt.title("imaginary time evolution: energy expectation")
    plt.show()
    # difference to ground state energy
    plt.semilogy(dτ*np.arange(nsteps), ϵ_traj - ϵmin)
    plt.ylabel(r"$\langle \psi(\tau) | H | \psi(\tau) \rangle - \epsilon_{\min}$")
    plt.title("imaginary time evolution: difference to exact ground state energy")
    plt.show()
    print("distance from the real ground:", ϵ_traj[nsteps-1] - ϵmin)
    
    return ϵ_traj[nsteps-1], ψ
```

```python
#First, calculate the ground state of the Hamiltonian with superconducting terms c*c + cc*
#the sup_normal below is used to add the superconducting terms
H_ini = H_Quad_topo((2,2),0,5) + sup_normal((2,2))
#start from |00000...>
ψ = np.zeros(2**16)
ψ[0] = 1
e_ini, v_ini = imaginary_time_evolution(ψ, H_ini)
```

```python
#do adiabatic

def adiabatic(H_initial, H_target, ϵ_tar, ψ_ini, dτ, nsteps):
     
    ϵ_traj = np.zeros(nsteps)
    ϵ_traj[0] = np.vdot(ψ_ini, H_target.dot(ψ_ini)).real
    ψ = ψ_ini  
    sup_term = sup_normal((2,2))
    
    for n in range(1, nsteps):
        
        s = n*(1/(nsteps-1))
        #set the Hamiltonian which vary with time 
        H = H_target + (1-s)*sup_term
        
        #extra time for final steps 
        if nsteps - n <2000:
            ψ = spla.expm_multiply(-(1j)*(dτ)*H, ψ)
        
        ψ = spla.expm_multiply(-(1j)*(dτ)*H, ψ)
        ψ /= np.linalg.norm(ψ)
        ϵ_traj[n] = np.vdot(ψ, H_target.dot(ψ)).real
        
        
    print(ϵ_traj[nsteps-1])
    
    plt.plot(dτ*np.arange(nsteps), ϵ_traj)
    
    plt.plot([0, nsteps*dτ], [ϵ_tar, ϵ_tar], "--")
    
    plt.xlabel("τ")
    plt.ylabel(r"$\langle \psi(\tau) | H_target | \psi(\tau) \rangle$")
    plt.title("adiabatic quantum computing: energy calculated by H_target")
    plt.show()
    
    plt.semilogy(dτ*np.arange(nsteps), ϵ_traj - ϵ_tar)
    plt.ylabel(r"$\langle \psi(\tau) | H_target | \psi(\tau) \rangle - \epsilon_{\min}$")
    plt.title("adiabatic quantum computing: difference to target state energy")
    plt.show()
    print("distance from the real ground:", ϵ_traj[nsteps-1] - ϵ_tar)
    #print("distance from the initial state:", ϵ_traj[nsteps-1] - εmin)
    
    
    return ϵ_traj[nsteps-1], ψ
```

```python
H_tar = H_Quad_topo((2,2),0, 5)
ϵ_tar ,v_tar = sparse.linalg.eigsh(H_tar, which = 'SA', k =1)
e_adia ,v_adia = adiabatic(H_ini, H_tar, ϵ_tar, v_ini, 0.2, 10001)
#to save time, you can run:
#e_adia ,v_adia = adiabatic(H_ini, H_tar, ϵ_tar, v_ini, 0.1, 101)
#to see whether it can work, but the final result will be inaccurate
```

```python
#define the number_op of four corners:
num1 = number_op(16, 1)
num4 = number_op(16, 2**3)
num12 = number_op(16, 2**11)
num16 = number_op(16, 2**15)
```

```python
#for our state obtained from adiabatic quantum computing, the <n> is 0.5 for each corner
print(np.vdot(v_adia, num1.dot(v_adia)))
print(np.vdot(v_adia, num4.dot(v_adia)))
print(np.vdot(v_adia, num12.dot(v_adia)))
print(np.vdot(v_adia, num16.dot(v_adia)))
#and such a state may not be suitable to do a quench:
```

```python
#H: Hamiltonian used to quench
#v: starting state
#num_nor: number operator. Please pick one of number operator defined above.

def quench(H, v, dτ, nsteps, num_nor):
    
    num_traj = np.zeros(nsteps)
    num_traj[0] = np.vdot(v, num_nor.dot(v)).real
    
    energy_traj = np.zeros(nsteps)
    energy_traj[0] = np.vdot(v, H.dot(v)).real
    
    for n in range(1, nsteps):
        v = spla.expm_multiply(-(1j)*(dτ)*H, v)
        v /= np.linalg.norm(v)
        num_traj[n] = np.vdot(v, num_nor.dot(v)).real
        energy_traj[n] = np.vdot(v, H.dot(v)).real
        
    plt.plot(dτ*np.arange(nsteps), num_traj)
    plt.show()
```

```python
#make use of the state obtained from adiabatic quantum computing to do a quench
#we change gamma = 0 to gamma = 1 in the Hamiltonian, while keep lambda = 5. 
H_quench = H_Quad_topo((2,2),1,5)
quench(H_quench, v_adia, 0.1, 301, num1)
# <n> almost keep unchange at <n> = 0.5 (notice y axis is 1e-13)
#maybe it is because the state used to do quench is a superposition of several ground states.
```
